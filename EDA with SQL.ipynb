
### Task 1

##### Display the names of the unique cities for listings


```python
%sql select distinct(CITY), count(*) as total_properties from REDFIN group by CITY order by COUNT desc
```

### Task 2

##### Display the different property types


```python
%sql select distinct(PROPERTY_TYPE),count(*) as total from REDFIN group by PROPERTY_TYPE
```

### Task 3

##### Investigate "Other" and "Multi-family" property type


```python
%sql select * from REDFIN where (PROPERTY_TYPE like 'Other%') or (PROPERTY_TYPE like 'Multi-Family%')
```

### Task 4

##### Determine minimum and maximum prices


```python
%sql select min(price) as min_price, max(price) as max_price from REDFIN
```

### Task 5

##### Determine minimum and maximum square footage


```python
%sql select min(square_feet) as min_sq_ft, max(square_feet) as max_sq_ft from REDFIN
```

### Task 6

##### Determine oldest home and newest home


```python
%sql select min(YEAR_BUILT) as oldest, max(YEAR_BUILT) as newest from REDFIN 
```

```python
%sql select YEAR_BUILT , count(*) as total_properties from REDFIN group by YEAR_BUILT order by YEAR_BUILT desc
```

### Task 7

##### Determine ranges for bedrooms and bathrooms


```python
%sql select beds,count(*) as total_properties from REDFIN group by beds
```

```python
%sql select baths,count(*) as total_properties from REDFIN group by baths
```

### Task 8

##### Determine range of lot sizes


```python
%sql select LOT_SIZE as min_lot, count(*) as number_properties from REDFIN where LOT_SIZE = (select min(LOT_SIZE) from REDFIN) group by LOT_SIZE
```

```python
%sql select LOT_SIZE as max_lot, count(*) as number_properties from REDFIN where LOT_SIZE = (select max(LOT_SIZE) from REDFIN) group by LOT_SIZE
```
